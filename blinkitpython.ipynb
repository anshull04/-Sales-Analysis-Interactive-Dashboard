## *KPI's REQUIREMENT*

```python
#total sales
total_sales = df['Sales'].sum()

#Avg sales
avg_sales = df['Sales'].mean()

#no of item sold
no_of_items_sold = df['Sales'].count()

#avg rating
avg_rating = df['Rating'].mean()

#display
print(f"total sales: ${total_sales:,.0f}")
print(f"avg sales: ${avg_sales:,.0f}")
print(f"no of item sold: ${no_of_items_sold:,.0f}")
print(f"avg rating: ${avg_rating:,.0f}")
```

# *CHARTS REQUIREMENT*

#### *TOTAL SALES BY FAT CONTENT*

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum() 
py.pie(sales_by_fat, labels = sales_by_fat.index,
               autopct = '%.1f%%',
                startangle = 90)
py.title('sales by fat content')
py.axis('equal')
py.show()
```

## *TOTAL SALES BY ITEM TYPE*

```python
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)

py.figure(figsize=(10, 6))
bars = py.bar(sales_by_type.index, sales_by_type.values)

py.xticks(rotation=90)
py.xlabel('Item Type')
py.ylabel('Total Sales')
py.title('Total Sales by Item Type')

for bar in bars:
    py.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f'{bar.get_height():.0f}',
        ha='center',
        va='bottom',
        fontsize=8
    )

py.tight_layout()
py.show()
```

## *FAT CONTENT BY OUTLET FOR TOTAL SALES*

```python
grouped = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()
grouped = grouped[['Regular', 'Low Fat']]

ax = grouped.plot(kind='bar', figsize=(8, 5), title='Outlet Tier by Item Fat Content')
py.xlabel('Outlet Location Tier')
py.ylabel('Total Sales')
py.legend(title='Item Fat Content')
py.tight_layout()
py.show()
```

## *TOTAL SALES BY OUTLET ESTABLISHMENT*

```python
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

py.figure(figsize=(9, 5))
py.plot(sales_by_year.index, sales_by_year.values, marker='o', linestyle='-')

py.xlabel('Outlet Establishment Year')
py.ylabel('Total Sales')
py.title('Outlet Establishment')

for x, y in zip(sales_by_year.index, sales_by_year.values):
    py.text(x, y, f'{y:.0f}', ha='center', va='bottom', fontsize=8)

py.tight_layout()
py.show()
```

## *SALES BY OUTLET SIZE*

```python
sales_by_size = df.groupby('Outlet Size')['Sales'].sum()

py.figure(figsize = (4,4))
py.pie(sales_by_size, labels = sales_by_size.index, autopct = '%1.1f%%', startangle = 90)
py.title('outlet size')
py.tight_layout()
py.show()
```

## *SALESBY OUTLET LOCATIONS*

```python
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales', ascending=False)

py.figure(figsize=(8, 3))  # Smaller height, enough width
ax = sns.barplot(x='Sales', y='Outlet Location Type', data=sales_by_location)

py.title('Total Sales by Outlet Location Type')
py.xlabel('Total Sales')
py.ylabel('Outlet Location Type')

py.tight_layout()  # Ensures layout fits without scroll
py.show()
```
